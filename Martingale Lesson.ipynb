#How to Win at Roulette (almost) Every Time
Here’s an almost guaranteed way to win \$5 at a roulette table.  
It’s called a Martingale betting strategy and it’s simple:
You put \$5 on black.  If you win, you walk away.  If you lose, you double your bet.  
You keep doing this until you win.  You only have to win once and eventually you win back all your losses, plus \$5.

#The Martingale Betting System
The goal is to win an amount equal to the min_bet, by walking away if you win, or doubling your bet every time you lose. Let's see what happens when we try this at our simulated roulette table.  

```python
plt.plot(play_martingale(1000,verbose=True, win_probability=0.5))
plt.xlabel('Roulette Spins')
plt.ylabel('Bankroll')
plt.title('Playing the Martingale')
```

#Iterated Martingale
Now let's play with fire.  Don't walk away if you win; instead, keep going until you reach a target amount of winnings.  Here we start with \$1000 and try to win another \$1000, but we end up down.

```python
iterate_martingale(1000,1000,min_bet=5,max_bet=500,win_probability=18.0/38, verbose=True)
```
